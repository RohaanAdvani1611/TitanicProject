## 5.Missing Values Treatment

```python
eda.isnull()
```

```python
eda.isnull().values.any()
```

```python
eda.info()
```

```python
eda.isnull().sum()
```

Missing Data will not make any impact on the result if its percentage is less 1%, if
missing datas range within the range of 1-5% then it is somehow manageable; however
in case of 5-15% complex techniques are used for handling the problems of missing
data but if it exceeds from 15% then it will surely hinder the result achieved after
applying data mining techniques.

1. Percentage of null values in Age column = (177/891)*100 = 19.86%
2. Percentage of null values in Cabin column = (687/891)*100 = 77.10%
3. Cabin having such high % of null values will hinder calculations in our model thus must be dropped.
4. Percentage of null values in Embarked column = (2/891)*100 = 0.22%
5. Embarked having such low % of null values will not effect calculations in our model thus can be ignored.

### A.Deleting observations or variables

```python
eda.drop('Cabin', axis=1, inplace=True)
```

```python
eda.head()
```

Thus Cabin has been dropped successfully.

### B.Single Imputation

1. If the variable is numeric then replace the missing values with either mean, median or mode.
2. If the variable is otherwise generally normally distributed (and in particular does not have any skewness), we would choose mean.
3. If the data skewed, median imputation is suggested.
4. If the variable is categorical then we could replace the missing values with the most frequent occurring value in that variable, i.e the mode.
5. We use the fillna function of pandas to replace na values with the value of our interest and
6. inplace=True command makes the permanently changes the value in that dataframe

```python
eda.hist(column=['Age'], bins=10)
```

Looking at the histogram of the variable Age (non missing values), we see that it is skewed. Hence we can impute missing values with Median of nonmissing data.

```python
eda['Age']=eda['Age'].fillna(value=eda['Age'].median())
```

```python
eda.info()
```

Thus the Null values of Age have been filled successfully.

```python
eda.isnull().sum()
```

Thus Null values have been treated successfully.

## 6.Outliers 

### A. Outlier Detection

```python
eda.boxplot(column=['Age'])
```

```python
eda['Age'].describe()
```

```python
IQR_Age=eda['Age'].quantile(0.75)-eda['Age'].quantile(0.25)
print(IQR_Age)
```

```python
Upper_OutlierLimit_Age = eda['Age'].quantile(0.75) + 1.5*IQR_Age
Lower_OutlierLimit_Age = eda['Age'].quantile(0.25) - 1.5*IQR_Age
print(Upper_OutlierLimit_Age)
print(Lower_OutlierLimit_Age)
```

```python
OutlierValues_Age = eda[(eda['Age']>=Upper_OutlierLimit_Age) | (eda['Age']<=Lower_OutlierLimit_Age)]
```
